# Entornos de desarrollo
-------

## Introducción
Como con tantas otras cosas de Python, disponemos de multitud de entornos de trabajo con los que desarrollar nuestras aplicaciones. Cada uno de ellos tiene sus características propias que los hacen preferibles para ciertos tipos de proyectos. Vamos a repasar los más comunes visitandolos en orden de complejidad creciente.

## El interprete de Python
*Python himself*. Al ser un lenguaje interpretado necesitamos un interprete para ejecutar el codigo. Normalmente se usa pasandole un nombre de fichero como parámetro, pero si lo le pasamos nada se inicia en modo interactivo y podemos escribir comandos directamente e ir ejecutandolos uno a uno.



#### Ventajas
- Por definición, va a estar instalado en todos los sitios donde hay a Python
- Es muy ligero.
#### Inconvenientes
- Es muy incomodo de usar, no tiene ningún tipo de ayuda para el programador

## IPython
Abreviación de *Interactive Python*. Es un interprete pensado desde el principio para ser usado de manera interactiva.
Incluye historial de comandos, autocompletado y unos comandos especiales llamados "magias" (%reset, %timeit, %edit, !)



### Ventajas
- Sigue siendo vastante ligero.
- Se puede instalar como un paquete más de Python
- Incluye algunas ayudas que hacen más cómoda la programación

### Inconvenientes
- No tiene debugger
- Es incomodo trabajar con ficheros

## Jupyter notebook
Es lo que habeis estado usando hasta ahora. Se crea un servidor web que presenta una página al usuario donde puede introducir código en Python y ejecutarlo:

Las páginas se dividen en celdas, que pueden ser de texto (*Markdown*) o ejecutables:

- A las celdas de texto se les puede dar formato mediante el lenguaje [markdown](https://en.support.wordpress.com/markdown-quick-reference/) permitiendo incluir listas, tablas, imagenes, videos, etc.

- Las celdas de código se ejecutan al pulsar Shift + Enter y el resultado aparece justo debajo.



Se utiliza para programación especulativa, para ir probando cosas y ver inmediatamente los resultados. También se usa mucho para generar informes y en educación (vease), para crear material de soporte interactivo.

#### Ventajas
- Permite crear documentos casi al vuelo, donde los resultados se van mostrando de manera orgánica.
- Permite experimentar facilmente viendo los resultados inmediatamente y pudiendo corregir celdas individuales.
- Puede gestionar multiples ficheros y módulos.
- Tiene un sistema de extensiones que permiten añadir nuevas funcionalidades (corrector ortográfico, indice automático, etc)

#### Inconvenientes
- El autocompletado podria ser mejor
- El debugger es muy primitivo
- No hay test unitarios


## Spyder
Este es el IDE que se instala por defecto con Anaconda. Es el que más se parece al IDE de Matlab. Tiene un area donde podemos editar ficheros, el interprete donde se ejecutan nuestras acciones y un visor de variables con el que podemos inspecctionar las variables que hayamos creado.



#### Ventajas
- Tiene un debugger adecuado
- Tiene visor de variables
- La gestión de ficheros es buena

#### Desventajas
- No lanza tests unitarios

## PyCharm
PyCharm es un IDE desarrollado por [JetBrains](https://www.jetbrains.com/). Tiene una versión de pago y otra gratuita, pero incluso la gratuita es adecuada para desarrollo profesional.



Tiene toda la funcionalidad de *Spyder*, pero ademas permite gestionar los entornos virtuales, integra herramientas de control de código, incluye corrector ortografico y de estilo y detecta y lanza automáticamente los tests unitarios, entre otras cosas.

#### Ventajas
- Tiene el set más completo de ayudas para el programador
- Permite gestionar los entornos virtuales
- El debbuger y el visor de variables son muy versatiles
- Detecta y lanza automaticamente los tests unitarios.
- Integra herramientas de control de código (git, hg, etc)

#### Desventajas
- Está implementado en java y puede ser un demasiado pesado para ordenadores poco potentes.
